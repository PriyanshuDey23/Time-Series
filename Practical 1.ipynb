# Time Resampling

```python
df.head(20)
```

```python
# yearly start frequency
df.resample(rule='A').min()
# rule(A) basically means year end frequency, we wil  get the year end data
# like the summary of the whole year
# there is different different rule present like a,b
# min is an aggregate fuction
# we will get the minimum row for that particular year
```

```python
df.resample(rule='A').min().plot()
```

```python
df.resample(rule='A').min()['Open'].plot()
```

```python
# we will get the maximum row for that particular year
# Yearly end frequency
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max().plot()
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
# quaterly date (rule=QS)
# quaterly end frequency
df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['High'].plot()
```

```python
# quaterly start frequency
df.resample(rule='QS').min()
```

```python
df.resample(rule='QS')['High'].min().plot()
```

```python
# Bussiness Start year Frequency
df.resample(rule='BA').min()
```

```python
df.resample(rule='BA').min()['High'].plot()
```

```python
# Bussiness End year Frequency
df.resample(rule='BA').max()
```

```python
df.resample(rule='BA').max()['High'].plot()
```

```python
# Business Quaters(Beginning)
df.resample(rule='BQS').min()
```

```python
df.resample(rule='BQS').min()['High'].plot()
```

```python
## Business Quaters(End)
df.resample(rule='BQS').max()
```

```python
df.resample(rule='BQS').max()['High'].plot()
```

```python
# plotting , do some kind of plotting
df['Open'].resample(rule='BA').mean().plot(figsize=(12,4),kind='bar')
```

```python
# Monthly mean
df['Open'].resample(rule='M').mean()
```

```python
# Monthly max
df['Open'].resample(rule='M').max()
```

```python
df['Open'].resample(rule='M').max().plot()
```

```python
# Rolling:-Provide rolling window calculations.
# it is rolling down or stepping by 10 steps
df['High'].rolling(10).mean().head(20)
# I am finding rolling mean
# mean wil be calculated between the 2 column and go down for the given window size i.e. 10
```

```python
df['High'].rolling(11).mean().head(20)
```

```python
df['Open:30 days rolling']=df['Open'].rolling(30).mean()
```

```python
df.head(32)
```

```python
# plot open with Open:30 days rolling
df[['Open','Open:30 days rolling']].plot(figsize=(12,4))
# first few value of open 30 days will be missing because of nan, check indivisually
# orange line is basically smoothing version of blue
```
